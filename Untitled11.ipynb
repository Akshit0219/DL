```python
import tensorflow as tf
import numpy as np
print("tensor version",tf.__version__)
```

```python
scalar=tf.constant(7)
print(scalar)
```

```python
vector=tf.constant([1,2,3])
print(vector)
```

```python
matrix=tf.constant([[1,2,3],[4,5,6]])
print(matrix)
```

```python
a=tf.constant([20,30,40])
b=tf.constant([1,2,3])
print(a+b)
print(a-b)
```

```python
a=tf.constant([20,30,40])
b=tf.constant([1,2,3])
print(a*b)
print(a/b)
```

```python
a=tf.constant([[20,30,40],[2,3,5]])
print(a.dtype,a.shape,tf.rank(a))
```
